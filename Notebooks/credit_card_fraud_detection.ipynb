# FEATUTE ENGINEERING AND EXTRACTION

```python
from sklearn.preprocessing import StandardScaler

# Feature Engineering
# Create 'hour' feature
X_train['hour'] = (X_train['Time'] // 3600) % 24
X_test['hour'] = (X_test['Time'] // 3600) % 24

# Create 'is_zero_amount' feature
X_train['is_zero_amount'] = (X_train['Amount'] == 0).astype(int)
X_test['is_zero_amount'] = (X_test['Amount'] == 0).astype(int)

# Drop original 'Time' column
X_train = X_train.drop('Time', axis=1)
X_test = X_test.drop('Time', axis=1)

#Log Transformation (Fix skewness BEFORE scaling) ---
X_train['Amount'] = np.log1p(X_train['Amount'])
X_test['Amount'] = np.log1p(X_test['Amount'])

#  Unified Scaling
cols_to_scale = X_train.columns.drop('is_zero_amount')

scaler = StandardScaler()

X_train[cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])
X_test[cols_to_scale] = scaler.transform(X_test[cols_to_scale])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    penalty="l2",
    C=1.0,
    class_weight= {0:1,1:4},
    max_iter=1000
)
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, recall_score, precision_score

# TEST SET METRICS
y_prob = model.predict_proba(X_test)[:,1]

best_threshold = 0
best_recall = 0

thresholds = np.arange(0.1, 0.9, 0.01)

for t in thresholds:
    y_pred_temp = (y_prob > t).astype(int)

    recall = recall_score(y_test, y_pred_temp, zero_division=0)
    precision = precision_score(y_test,  y_pred_temp, zero_division=0)

    # keep recall high but avoid terrible precision
    if recall > best_recall and precision > 0.5:
        best_recall = recall
        best_threshold = t

print("Best Threshold:", best_threshold)
print("Best Recall:", best_recall)

# APPLY BEST THRESHOLD HERE
final_preds = (y_prob > best_threshold).astype(int)

# CONFUSION MATRIX
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, final_preds))

# CLASSIFICATION REPORT
print("\nClassification Report:")
print(classification_report(y_test, final_preds))
```

RANDOM FORSEST CLASSIFIER


```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=150,
    max_depth=10,
    min_samples_leaf=100,
    max_samples=0.1,
    max_features = 'sqrt',
    class_weight="balanced",   # recall-oriented
    random_state=42,
    n_jobs=-1
)
```

```python
from sklearn.model_selection import StratifiedKFold, cross_validate, cross_val_predict
from sklearn.metrics import fbeta_score, recall_score, precision_score, roc_auc_score, roc_curve

# Initialize Cross-Validation
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
metrics = ['recall', 'f1', 'roc_auc']

# Check general model performance
results = cross_validate(rf, X_train, y_train, cv=cv, scoring=metrics, n_jobs=-1)

print(f"--- CV Baseline Performance ---")
print(f"CV Recall: {results['test_recall'].mean():.4f}")
print(f"CV F1: {results['test_f1'].mean():.4f}")
print(f"CV ROC-AUC: {results['test_roc_auc'].mean():.4f}\n")

# Optimize Threshold using Out-of-Fold Training Probabilities
y_train_probs_rf = cross_val_predict(rf, X_train, y_train, cv=cv, method='predict_proba', n_jobs=-1)[:, 1]

thresholds = np.arange(0.01, 0.9, 0.01)
best_t_rf = 0.5
best_score = 0

for t in thresholds:
    y_train_pred_temp = (y_train_probs_rf >= t).astype(int)

    # fbeta_score with beta=2 weights Recall higher than Precision
    current_score = fbeta_score(y_train, y_train_pred_temp, beta=2)

    if current_score > best_score:
        best_score = current_score
        best_t_rf = t

print(f"Optimal Threshold (tuned for Recall): {best_t_rf:.2f}")

# Final Evaluation on the Test Set
rf.fit(X_train, y_train)

# Predict using the optimized threshold
y_test_probs_rf = rf.predict_proba(X_test)[:, 1]
y_test_pred_rf = (y_test_probs_rf >= best_t_rf).astype(int)

print(f"\n--- Final Test Results (Threshold: {best_t_rf:.2f}) ---")
print(f"Test Recall:    {recall_score(y_test, y_test_pred_rf):.4f}")
print(f"Test Precision: {precision_score(y_test, y_test_pred_rf):.4f}")
print(f"Test ROC-AUC:    {roc_auc_score(y_test, y_test_probs_rf):.4f}")
print(f"Test F-Beta(2): {fbeta_score(y_test, y_test_pred_rf, beta=2):.4f}")

# ROC CURVE
fpr, tpr, _ = roc_curve(y_test, y_test_probs_rf)

plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, y_test_probs_rf):.3f}")
plt.plot([0,1], [0,1], '--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

LIGHTGBM


```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (precision_score, recall_score, f1_score,
                             classification_report, roc_auc_score,
                             roc_curve, precision_recall_curve)
import lightgbm as lgb

# 1. Setup Cross-Validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_precisions, cv_recalls, cv_aucs, cv_thresholds = [], [], [], []

# 2. Iterative Training (Cross-Validation)
for train_idx, val_idx in skf.split(X_train, y_train):
    X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[val_idx]
    y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

    model = lgb.LGBMClassifier(
        n_estimators=1000,
        learning_rate=0.02,
        num_leaves=63,
        min_child_samples=20,
        subsample=0.8,
        colsample_bytree=0.8,
        scale_pos_weight=100,
        verbosity=-1,
        random_state=42,
        n_jobs=-1
    )

    model.fit(X_tr, y_tr)
    y_val_probs = model.predict_proba(X_val)[:, 1]

    # 3. Precision-Targeted Threshold Logic
    precisions, recalls, thresholds = precision_recall_curve(y_val, y_val_probs)

    # We look for the highest recall where precision is at least 75%
    target_p = 0.75

    # Calculate valid_indices for the current fold
    valid_indices = np.where(precisions >= target_p)[0]

    # F2-score (recall-focused) - This was being used before, but the logic below
    # seems to be targeting a specific precision. Keep it for now in case of fallback.
    f2_scores = (5 * precisions * recalls) / (4 * precisions + recalls + 1e-9)

    # Removed the initial best_idx = np.argmax(f2_scores[:-1]) as it conflicts
    # with the conditional logic that follows.

    if len(valid_indices) > 0:
        # Of those meeting precision, pick the one with max recall
        best_idx = valid_indices[np.argmax(recalls[valid_indices])]
        best_t = thresholds[min(best_idx, len(thresholds)-1)]
    else:
        # Fallback: If 75% isn't reachable, find the best F1-score threshold
        f1_scores = 2 * (precisions * recalls) / (precisions + recalls + 1e-9)
        best_t = thresholds[np.argmax(f1_scores[:-1])]

    y_val_pred = (y_val_probs >= best_t).astype(int)

    cv_precisions.append(precision_score(y_val, y_val_pred, zero_division=0))
    cv_recalls.append(recall_score(y_val, y_val_pred, zero_division=0))
    cv_aucs.append(roc_auc_score(y_val, y_val_probs))
    cv_thresholds.append(best_t)

# 4. Final Evaluation on Test Set
final_threshold = np.median(cv_thresholds)

# Re-fit on full training data
model.fit(X_train, y_train)
y_probs = model.predict_proba(X_test)[:, 1]
y_pred = (y_probs >= final_threshold).astype(int)

# --- Output Results ---
print(f"--- LightGBM CV Metrics ---")
print(f"Mean CV Precision: {np.mean(cv_precisions):.4f}")
print(f"Mean CV Recall:    {np.mean(cv_recalls):.4f}")
print(f"Mean CV ROC-AUC:   {np.mean(cv_aucs):.4f}")
print(f"Used Threshold:    {final_threshold:.4f}")

print("\n--- Final Test Report ---")
print(classification_report(y_test, y_pred))

# Save these for your comparison table
test_precision_lgbm = precision_score(y_test, y_pred)
test_recall_lgbm = recall_score(y_test, y_pred)
test_f1_lgbm = f1_score(y_test, y_pred)
test_auc_lgbm = roc_auc_score(y_test, y_probs)

# 5. Visualize ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_probs)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"LightGBM (AUC = {test_auc_lgbm:.3f})")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("LightGBM ROC Curve")
plt.legend()
plt.show()
```

ENSEMBLE (LightGBM + CatBoost)

```python
from sklearn.metrics import classification_report, roc_auc_score
import lightgbm as lgb
from catboost import CatBoostClassifier
import numpy as np

# 1. Initialize LightGBM with your tuned parameters
lgbm_model = lgb.LGBMClassifier(
    n_estimators=1000,
    learning_rate=0.02,
    num_leaves=63,
    scale_pos_weight=100,
    verbosity=-1,
    random_state=42
)

# 2. Initialize CatBoost with your tuned parameters
cb_model = CatBoostClassifier(
    iterations=400,
    learning_rate=0.03,
    depth=6,
    auto_class_weights='Balanced',
    verbose=False,
    random_state=42
)

# 3. Train both models
lgbm_model.fit(X_train, y_train)
cb_model.fit(X_train, y_train)

# 4. Get probabilities for the positive class (Class 1)
lgbm_probs = lgbm_model.predict_proba(X_test)[:, 1]
cb_probs = cb_model.predict_proba(X_test)[:, 1]

# 5. Ensemble: Weighted Average
# CatBoost showed higher precision in your comparison table,
# so we give it slightly more weight.
ensemble_probs = (0.4 * lgbm_probs) + (0.6 * cb_probs)

# 6. Apply Threshold and Evaluate
# Using 0.5 as a starting point; you can tune this based on your PR curve
threshold = 0.5
ensemble_preds = (ensemble_probs >= threshold).astype(int)

print("--- Ensemble (LGBM + CatBoost) Performance ---")
print(classification_report(y_test, ensemble_preds))
print(f"Ensemble ROC-AUC Score: {roc_auc_score(y_test, ensemble_probs):.4f}")
```
